1、打开文件

```python
import csv
```

```python
iris_data = []
```

```python
with open("./iris.csv") as csvfile:
    csv_reader = csv.reader(csvfile)
    birth_header = next(csv_reader)
    for row in csv_reader: 
        iris_data.append(row)
```

```python
print(iris_data)
```

2、创建numpy的array数组iris_array，每行内容是将iris_data中的每行数据的第一列去掉；

```python
import numpy as np
```

```python
iris_array=np.array(iris_data)[:,1:]
```

```python
print(iris_array)
```

3、查看iris_array的类型；

```python
print(type(iris_array))
```

4、查看iris_array的shape；

```python
print(iris_array.shape)
```

5、将iris_array的第五列去掉并将数据类型转为float，赋值给新的array数组new_array；

```python
del_array=np.delete(iris_array,4,axis=1)
```

```python
new_array=del_array.astype(np.float32)
```

```python
print(new_array)
```

6、对new_array的前5行按行排序；

```python
array_sort = np.sort(new_array[:5],axis=1)
```

```python
print(array_sort)
```

7、对new_array的前5行按行求和；

```python
array_sum = np.sum(new_array[:5],axis=1)
```

```python
print(array_sum)
```

8、求new_array的第一列的平均值；

```python
array_mean = np.mean(new_array[:,0],axis=0)
```

```python
print(array_mean)
```

9、求new_array的第一列的方差；

```python
array_variance = np.std(new_array[:,0],axis=0)
```

```python
print(array_variance)
```
